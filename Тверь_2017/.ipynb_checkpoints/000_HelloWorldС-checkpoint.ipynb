# Знай  Славянский Модифицированный Диалект (CMD)

```python
:: Start.Bat
:: отключение предварительног вывода команды в консоль перед исполнением
@echo off
:: Переход в папку, к которой находится исполняемый батник
cd %~dp0
:: Находим в недрах С-Free компилятор и шлем свой файлик туда.
:: Помечаем пораметром название файла после компиляции
"C:\Program Files (x86)\C-Free 5\mingw\bin\c++.exe" main.c -o Start.exe
:: запускаем результат компиляции
Start.exe
:: 5 секунд радуемся осторожненько
TIMEOUT /T 5
:: аккуратно выключаем батничеГ.
exit /b        
```

Теперь кодим в <a href="https://notepad-plus-plus.org/">Notepad++</a>. Сохраняем через сочетание клавиш <Code>Ctrl+S</Code>. Запускаем через <Code>F5; Enter</Code>, в качестве параметров запуска подсовываем только что написанный батник.

Пользуемся правильной кодировкой файла.</br>
Для батника это : <code>Кодировки/Кодировки/Кирилица/OEM866</code> </br>
Для main.c файла это : <code>Кодировки/Кодировки/Кирилица/Windows-1257</code> </br>
